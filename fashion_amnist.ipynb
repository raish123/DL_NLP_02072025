```python
#importing pytorch module which will be used in this project!!
import torch
import  torch.nn as nn
import pandas as pd
import sklearn
from torch.utils.data import DataLoader,Dataset
```

```python
#now checking cuda or gpu is available in my system or not
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
#now loading the dataset by using pandas library
data = pd.read_csv("fmnist_small.csv")
data.head()
```

```python
#Dropping those duplicates record
data.drop_duplicates(inplace=True)
```

```python
#selecting input and output variable
x = data.drop(columns=['label'],axis=1)
y = data['label']
```

```python
#splitting the dataset into train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
type(x_train)
```

```python
#now performing input scaling by using min,max scaling
x_train = x_train/255.0
x_test = x_test/255.0
```

# now changing thise training and testing set to numpy array object

```python
import numpy as np
```

```python
x_train = np.array(x_train,dtype=np.float32)
x_test = np.array(x_test,dtype=np.float32)
y_train = np.array(y_train,dtype=np.int64)
y_test = np.array(y_test,dtype=np.int64)
```

```python
x_train.ndim,x_train.dtype
```

# now changing this numpy object to tensor object

```python
x_train = torch.from_numpy(x_train)
x_test = torch.from_numpy(x_test)
y_train = torch.from_numpy(y_train)
y_test = torch.from_numpy(y_test)
```

```python
y_train = y_train.long()   # Convert float to long (because we have to change if we have multi class output)
y_test = y_test.long()
```

# now creating dataloader object by using custom dataset class

```python
#In this class we define feature and label and also showing shape and returning based on index feature and label
class CustomDataset(Dataset): 
    #in this class we used 3 special or magical method
    def __init__(self,feature,label):
        super().__init__()
        self.feature = feature
        self.label = label
        
    def __len__(self):
        return len(self.feature)
    
    def __getitem__(self, index):
        return self.feature[index],self.label[index]
        
```

```python
#creating an object of CustomDataset class
training_dataset = CustomDataset(x_train,y_train)
testing_dataset = CustomDataset(x_test,y_test)
```

```python
#now creating the object of DataLoader class
train_dataloader =   DataLoader(training_dataset,batch_size=10,shuffle=True,pin_memory=True)
testing_dataloader = DataLoader(testing_dataset,batch_size=10,shuffle=False,pin_memory=True)
```

# creating a custom neural network architecture inheriting parent class property Module

```python
class BasicArchitecture(nn.Module):
    #using constructor method for initializing the trainable parameter(weights and bias)
    def __init__(self,input_feature):
        
        super().__init__()
        
        #defining trainable parameter of first layer
        self.linear1 = nn.Linear(input_feature,128)
        self.relu1 = nn.ReLU()
        
        #defining 2nd layer of trainable parameter
        self.linear2 = nn.Linear(128,64)
        self.relu2 = nn.ReLU()
        
        #defining 3rd layer of trainable parameter
        self.linear3 = nn.Linear(64,10)
        
        
    #defining the forward propogation algorithm!!
    def forward(self,input_feature):
        #output of first second and third layer
        output = self.relu1(self.linear1(input_feature))
        output = self.relu2(self.linear2(output))
        output = self.linear3(output)
        
        return output
```

```python
x_train.shape[1]
```

```python
#creating an object of BasicArchitecture class
model = BasicArchitecture(input_feature=x_train.shape[1]).to(device)
model
```

```python
#if i want to see the summary of the model
from torchinfo import summary
summary(model)
```

```python
#now defining the optimizers or compiling model 
lr = 0.01
epochs = 50

optimizer = torch.optim.Adam(model.parameters(),lr=lr) 
#optimizer will reduce the loss gradulally and helping to reach to global minima point
optimizer
```

```python
for feature,label in train_dataloader:
    
    print(feature)
    print(label)
```

```python
loss_function = nn.CrossEntropyLoss() #this will calculate loss for sparse categorical cross entropy
```

```python
#want to see what will be initialize wts and bias in each layer
model.linear1.weight
```

```python
model.linear1.bias
```

# now training the model

```python
#now defining the training steps
for epoch in range(epochs):
    total_epoch_loss = 0
    
    for feature,label in train_dataloader:
        #loading batch data into gpu
        feature = feature.to(device)
        label = label.to(device)
        #now calling the forward propgation method
        y_pred = model(feature)
        
        #calculating loss function
        loss  =loss_function(y_pred,label)
        
        #now clearing the optimizer object gradient
        optimizer.zero_grad()
        
        
        #now finding out the loss of gradient wrt to parameter by using back propogation algorithm
        loss.backward()
        
        #now updating wts and bias in each layer using optimizer step method
        optimizer.step()
        
        #batch wise loss calculating
        total_epoch_loss = total_epoch_loss+loss.item()
    avg_loss = total_epoch_loss/(len(train_dataloader))
        
    print(f"epochs is :{epoch+1} and loss will be {avg_loss}")
        
```

# evaluating the model

```python
model.eval() #this line we externally added to start evaluation of model
```

```python
correct = 0
total = 0
with torch.no_grad():
    #passing the test dataset loadr to the model
    for batch_feature,batch_label in testing_dataloader:
        batch_feature = batch_feature.to(device)
        batch_label = batch_label.to(device)
        
        #calling the forward pass
        y_pred = model(batch_feature)
        
        _, predicted = torch.max(y_pred,1) #rtn label wrt to max probability
        
        total = total+batch_label.shape[0]
        
        correct = correct + (predicted==batch_label).sum().item()
        
        
    accuracy = 100 * correct / total
print(f"\nâœ… Test Accuracy: {accuracy:.2f}%")
```
